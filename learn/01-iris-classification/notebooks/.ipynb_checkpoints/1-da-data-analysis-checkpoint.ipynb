### Environment Setup
Lets first setup our environment.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score 
from sklearn.neighbors import KNeighborsClassifier

import warnings
warnings.filterwarnings('ignore') # supress warnings
```

### Data Preview
Load the data, give headers if needed. Preview the data statistics.

```python
data_dir = 'data/iris.data'
data = pd.read_csv(data_dir, names=['sepalWidth', 'sepalLength', 'petalLength', 'petalWidth', 'class'])
data.head()
```

```python
data.describe()
```

### Data Visualization
Visualize data with proper plots, Seaborn's Pair Plot is used here.

```python
sns.pairplot(data, hue='class', markers='.')
```

### Model Optimizing
The small dataset allows us to simply use a KNN(*K-Nearest Neighbors*) model. The `k` constant has to be selected for accuracy readings, this case simply by testing all possible `k` values (1 to data width). The results are then plotted. 

```python
X = data.drop(['class'], axis=1)
y = data['class']
da
```

```python
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state=1, test_size=0.4)
k_range = list(range(1,train_X.shape[0]))
scores = []
for k in k_range:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(train_X, train_y)
    pred_y = knn.predict(val_X)
    scores.append(accuracy_score(val_y, pred_y))

plt.plot(k_range, scores)
plt.xlabel('KNN k constant')
plt.ylabel('Accuracy')
plt.title('Accuracy against k constant')
plt.show()
```

It seems that `k = 12` is a good value to use for this dataset.

### Model Fitting
Now we can recreate our model with the proper `k` value. We now can use a larger `test_size`.

```python
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state=1, test_size=0.6)
model = KNeighborsClassifier(n_neighbors=12)
model.fit(train_X, train_y)
pred_y = model.predict(val_X)
print('Accuracy: {:f}'.format(accuracy_score(val_y, pred_y)))
```
